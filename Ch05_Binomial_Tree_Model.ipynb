## 1. Binomial Tree Model

The **Binomial Tree Model** is a discrete-time framework used to price derivatives, especially options. It assumes that over each small time step, the underlying asset price can move to one of two possible values.

$$
S_n \to 
\begin{cases}
S_{n+1}^u & \text{(up move)} \\
S_{n+1}^d & \text{(down move)}
\end{cases}
$$

<span style="color:red">
    
There are indeed two common ways to write the up/down movement.

**(1) Percentage Move Form** (used in "Mathematics for Finance: An Introduction to Financial Engineering" by Marek Capinski & Tomasz Zastawniak)
$$
S_{n+1}^u = S_{n}(1+u), \;\; S_{n+1}^d = S_{n}(1+d)
$$

where,
- $u$ and $d$ are percentage changes
- $0<u<1$
- $-1<d<0$

**(2) Multiplicative Move Form (Up/Down Factor)** (used in "Options, Futures and Other Derivatives" by John C. Hull)
$$
S_{n+1}^u = S_{n}u, \;\; S_{n+1}^d = S_{n}d
$$

where,
- $u$ and $d$ are 
- $0 < d < u$

</span>

In practice:

Hull (Options, Futures, and Other Derivatives) is more standard in option pricing, especially when:
- defining risk-neutral probability
- Building recombining trees

**This Notebook will also follow the Multiplicative Move Form.**

### Set up a Binomial Tree

Time Structure

- Consider a time horizon $[0,T]$, $t \in [0,T]$
- Divide the time horizon into $n$ equal intervals of length $h$: $T=nh$

Underlying Asset

- Let $S(t)$ denote the asset price at time $t$
- $S(t) \ge 0$
- Define the discrete-time process:
  $S(n)=S(t_n)$, where the stock price at step $n$ corresponds to the price at time $t=nh$



### One-Step Binomial Model

Let $S_1$ be a random variable with:
  $\Omega = \{\text{up}, \text{down}\}$,  
  $S_1 : \Omega \to [0, \infty)$

Let the multiplicative factor be:

  $K^\omega =
    \begin{cases}
      u & \text{if } \omega = \text{up} \\
      d & \text{if } \omega = \text{down}
    \end{cases}$

where:
  $u, d \in \mathbb{R}$, with $0 < d < u$

The next-step price is:
  $S_1^\omega = S_0 \cdot K^\omega$

Explicitly:

  $S_1 =
    \begin{cases}
      S_0 \cdot u & \text{(up)} \\
      S_0 \cdot d & \text{(down)}
    \end{cases}$

### Two-step Binomial Model

Sample space:
  $\Omega = \{\text{uu}, \text{ud}, \text{du}, \text{dd}\}$

At each step, the stock evolves as:
  $S_{n+1} = S_n \cdot K_{n+1}$

where:
  $K_{n+1} =
    \begin{cases}
      u & \text{(up)} \\
      d & \text{(down)}
    \end{cases}$

with $0 < d < u$

Starting from $S_0$, after two steps:$S_2 = S_0 \cdot K_1 \cdot K_2$

$$
S_2 =
\begin{cases}
S_0 \cdot u^2 & \omega = \text{uu} \\
S_0 \cdot u d & \omega = \text{ud} \\
S_0 \cdot d u & \omega = \text{du} \\
S_0 \cdot d^2 & \omega = \text{dd}
\end{cases}
$$


Since multiplication is commutative:
  
  $u d = d u$

Therefore:
  
  $S_{ud} = S_{du} = S_0 \cdot u d$

### General Binomial Model

Consider a Binomial Model with n steps:

Sample space: $\Omega = \{\text{up}, \text{down}\}^n$

A path (a sequence of up/down moves) is denoted as $\omega$, where $\omega \in \Omega$.

- $\omega_k$ = k-th letter of $\omega$
- In total, there are $2^n$ paths, $\mathcal{F}=2^n$

Price formula: $S_n = S_0 \cdot u^k \cdot d^{(n-k)}$, where $k = $the number of ups.

Probability of k up moves: $p^k \cdot (1-p)^{(n-k)}$, where $p = $the probability of up.

## 2. Using Binomial Tree in Option Pricing

In the binomial tree framework, there are three fundamental approaches to pricing a derivative (such as an option). Although they appear different in methodology, all three are based on the same underlying principle: **the absence of arbitrage**.

This section introduces the three approaches: Replicating portfolio, No-arbitrage, and Risk-neutral.

<span style="color:blue">

### Option Pricing Example 

Consider a European call option written on a stock in a two-step binomial model.

Given:

- Current stock price:
  $S_0 = 100$

- Strike price:
  $K = 100$

- Up factor:
  $u = 1.1$

- Down factor:
  $d = 0.9$

- Risk-free interest rate per period:
  $r = 0.05$

- Time:
  $T = 2$years, Steps $N = 2$

Thus, the possible stock prices are:

- At time 0:
  $S_0 = 100$

- At time 1:
  $S_u = 100(1.1) = 110$
  
  $S_d = 100(0.8) = 90$

- At time 2:
  $S_{uu} = 100(1.1)^2 = 121$
  
  $S_{ud} = 100(1.1)(0.8) = 99$
  
  $S_{dd} = 100(0.8)^2 = 81$

Option Payoff at Maturity

- At node $uu$:
  $C_{uu} = \max(121 - 100, 0) = 21$

- At node $ud$:
  $C_{ud} = \max(99 - 100, 0) = 0$

- At node $dd$:
  $C_{dd} = \max(81 - 100, 0) = 0$

### Replicating Portfolio Approach

- Construct a portfolio of the underlying asset and a risk-free bond
- The portfolio replicates the payoff of the option in all possible future states
- By the law of one price, the cost of this portfolio determines the option price

For the sample problem above,

Consider that we construct, at each node, a portfolio of:

- $\Delta$ shares of stock
- $B$ units of the risk-free assets (bond)

The value of the portfolio: $\Pi = \Delta S + B$

We want the portfolio to match the option payoff in the next period.

At the up node $S_u$, the stock price is $S_u = 110$. 

The next-step option payoffs are:
  $C_{t=2} =
    \begin{cases}
      C_{uu} = 21 & \text{if } \omega = \text{up} \\
      C_{ud} = 0 & \text{if } \omega = \text{down}
    \end{cases}$

Construct a replicating portfolio:

$$
\Delta_u \cdot 121 + B_u \cdot (1+0.05) = 21
\Delta_u \cdot 99 + B_u \cdot (1+0.05) = 0
$$

Solve for:

$\begin{cases}
    \Delta_u = 0.9545 \\
    B_u = -90
\end{cases}$

$C_u = 0.9545 \cdot 110 + (-90) = 15$

The replicating portfolio consists of 0.9545 units of stock and a 90-unit borrow at the risk-free rate. The value of the portfolio (the option price) at the up node is $C_u = 15$.

At the down node $S_d$, the stock price is $S_u = 90$. 

The next-step option payoffs are:
  $C_{t=2} =
    \begin{cases}
      C_{ud} = 0 & \text{if } \omega = \text{up} \\
      C_{dd} = 0 & \text{if } \omega = \text{down}
    \end{cases}$

Construct a replicating portfolio:

$$
\begin{cases}
\Delta_u \cdot 99 + B_u \cdot (1+0.05) = 0 \\
\Delta_u \cdot 81 + B_u \cdot (1+0.05) = 0
\end{cases}
$$

Solve for:

$\begin{cases}
    \Delta_d = 0 \\
    B_d = 0
\end{cases}$

$C_d = 0$

Now replicate the option using the values at time 1:

- up state: $C_u = 15$
- down state: $C_d = 0$

Construct a portfolio

$$
\Pi_0 = \Delta_0 S + B_0
$$

such that 

$$
\begin{cases}
\Delta_0 \cdot 110 + B_0(1.05) = 15 \\
\Delta_0 \cdot 90 + B_0(1.05) = 0
\end{cases}
$$

Solve for:

$\begin{cases}
    \Delta_0 = 0.75 \\
    B_0 = -64.2857
\end{cases}$

$C_0 = 0.75 \cdot 100 + (-64.2857) = 10.7143$

The price of the European call option is $C_0=10.7143$. The replicating portfolio consists of 0.75 units of stock and borrowing 64.2857 at the risk-free rate.

### No-Arbitrage Approach

- Assume that arbitrage opportunities do not exist in the market
- A portfolio of options and stocks should have the same value at the end
- This approach provides the fundamental justification for pricing

For the sample problem above, 

Consider that we construct, at each node, a hedged portfolio of:

- long $\Delta$ shares of stock
- short 1 call option

The value of the hedged portfolio is

$$
\Pi = \Delta S - C
$$

We choose $\Delta$ so that the portfolio has the same value in the up and down states in the next period.  
Then the portfolio is risk-free, so by no-arbitrage, it must grow at the risk-free rate.

At the up node $S_u$, the stock price is $S_u = 110$.

The next-step option payoffs are:

$$
C_{t=2} =
\begin{cases}
C_{uu} = 21 & \text{if } \omega = \text{up} \\
C_{ud} = 0 & \text{if } \omega = \text{down}
\end{cases}
$$

Construct a hedged portfolio:

$$
\begin{cases}
\Delta_u \cdot 121 - 21 = \Pi_{uu} \\
\Delta_u \cdot 99 - 0 = \Pi_{ud}
\end{cases}
$$

Set the two future portfolio values equal:

$$
\Delta_u \cdot 121 - 21 = \Delta_u \cdot 99
$$

Solve for:

$$
\Delta_u = \frac{21}{22} = 0.9545
$$

Then the portfolio payoff in the next period is

$$
\Pi_{uu} = \Pi_{ud} = 0.9545 \cdot 99 = 94.5
$$

Discount back one step:

$$
\Pi_u = \frac{94.5}{1.05} = 90
$$

Now use the current portfolio value:

$$
\Pi_u = \Delta_u \cdot 110 - C_u
$$

$$
90 = 0.9545 \cdot 110 - C_u
$$

$$
C_u = 15
$$

The hedge portfolio consists of 0.9545 units of stock and a short 1 call.  
The value of the call option at the up node is $C_u = 15$.

At the down node $S_d$, the stock price is $S_d = 90$.

The next-step option payoffs are:

$$
C_{t=2} =
\begin{cases}
C_{du} = 0 & \text{if } \omega = \text{up} \\
C_{dd} = 0 & \text{if } \omega = \text{down}
\end{cases}
$$

Construct a hedged portfolio:

$$
\begin{cases}
\Delta_d \cdot 99 - 0 = \Pi_{du} \\
\Delta_d \cdot 81 - 0 = \Pi_{dd}
\end{cases}
$$

Set the two future portfolio values equal:

$$
\Delta_d \cdot 99 = \Delta_d \cdot 81
$$

Solve for:

$$
\Delta_d = 0
$$

Then

$$
\Pi_{du} = \Pi_{dd} = 0
$$

Discount back one step:

$$
\Pi_d = 0
$$

Now use the current portfolio value:

$$
\Pi_d = \Delta_d \cdot 90 - C_d
$$

$$
0 = 0 \cdot 90 - C_d
$$

$$
C_d = 0
$$

The hedge portfolio at the down node is trivial, and the value of the call option is $C_d = 0$.

Now go back to time 0 using the values at time 1:

- up state: $C_u = 15$
- down state: $C_d = 0$

Construct a hedged portfolio

$$
\Pi_0 = \Delta_0 S_0 - C_0
$$

such that the future portfolio values are equal:

$$
\begin{cases}
\Delta_0 \cdot 110 - 15 = \Pi_u \\
\Delta_0 \cdot 90 - 0 = \Pi_d
\end{cases}
$$

Set them equal:

$$
\Delta_0 \cdot 110 - 15 = \Delta_0 \cdot 90
$$

Solve for:

$$
\Delta_0 = \frac{15}{20} = 0.75
$$

Then the riskless payoff at time 1 is

$$
\Pi_u = \Pi_d = 0.75 \cdot 90 = 67.5
$$

Discount back to time 0:

$$
\Pi_0 = \frac{67.5}{1.05} = 64.2857
$$

Now use the current portfolio value:

$$
\Pi_0 = \Delta_0 \cdot 100 - C_0
$$

$$
64.2857 = 0.75 \cdot 100 - C_0
$$

$$
C_0 = 10.7143
$$

The price of the European call option is $C_0 = 10.7143$.

The hedged portfolio at time 0 consists of 0.75 units of stock and short 1 call option.

### Risk-Neutral Valuation

**Risk-Neutral Probability**

In reality, 

- The stock has some unknown true probability of going up/down
- Investors require a risk premium

Therefore, 

$E[S_{t+1}] \ge S_t e^r$

Here, we define a fake probability $p^*$ such that:

$E^*[S_{t+1}] = S_t e^r$

under this scenario, 

- all assets grow at the risk-free rate
- investors are "as if" risk-neutral

Consider a one-step binomial model, where $f$ is the value of the option.

<img src="Pictures/One_Step_Binomial_Model.png" width="400">

We can build a portfolio consisting of 

- a long position on $\Delta$ shares of the stock at time 0
- a short position on one option of the stock at time 0

The present value of the portfolio: 
$\Pi_0 = S_0 \Delta - f$

At maturity,

$$
\begin{cases}
\Pi_u = \Delta S_u - f_u \\
\Pi_d = \Delta S_d - f_d \\
\end{cases}
$$

To eliminate risk, choose $\Delta$ such that
$\Pi_u = \Pi_d$

so,

$\Delta = \frac{f_u-f_d}{S_u-S_d}$

Because the problem is riskless, it must grow at the risk-free rate $r$

$\Pi_0 e^{-rT} = \Pi_u = \Pi_d$

Subsitue $\Pi_0$ and $\Pi_u$,

$\Delta S_0 -f = e^{-rT} (\Delta S_u - f_u)$

Subsitue $\Delta$ and solve,

$\boxed{f = e^{-rT}(p^* f_u + (1-p^*) f_d)}$

$p^* = \frac{e^{rT}S_0 - S_d}{S_u - S_d}$

Subsitue $S_u = S_0 u$ and $S_d = S_0 d$,

$\boxed{p^* = \frac{e^{rT} - d}{u - d}}$

or discrete compounding

$p^* = \frac{1+r-d}{u-d}$

**Back to our sample option**

Given the information, 
$p^* = \frac{1 + 0.05 - 0.9}{1.1 - 0.9} = 0.75$

At the up node $S_u$, the next step payoffs are:

$$
C_{t=2} =
\begin{cases}
C_{uu} = 21 & \text{if } \omega = \text{up} \\
C_{ud} = 0 & \text{if } \omega = \text{down}
\end{cases}
$$

The value of the option $C_u$:

$C_u = \frac{1}{1+0.05} [0.75 \cdot 21 + (1-0.75) \cdot 0] = 15$

At the down node $S_d$, the value of the option $C_d = 0$

Now go back to time 0:

$C_0 = \frac{1}{1+0.05} [0.75 \cdot 15 + (1-0.75) \cdot 0] = 10.7143$
