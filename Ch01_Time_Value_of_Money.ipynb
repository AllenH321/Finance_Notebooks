## 1. Present Value & Future Value

$$
\begin{aligned}
\text{Simple Interest:} &\quad V(t) = P(1+rt) \\
\text{Discrete Compound Interest:} &\quad V(t) = P\left(1+\frac{r}{m}\right)^{mt} \\
\text{Continuous Compound Interest:} &\quad V(t) = Pe^{rt}
\end{aligned}
$$

---


## 2. Effective Rate of Interest

For a given compounding method with interest rate $r$, the eﬀective rate $r_e$ gives the same growth factor over one year under annual compounding.

Converting $r$ in different compounding methods into $r_e$ helps compare the profitability of different methods.

$$
\begin{aligned}
\text{Simple Interest:} &\quad r_e = r \\
\text{Discrete Compound Interest:} &\quad r_e = (1+r/m)^{m} - 1 \\
\text{Continuous Compound Interest:} &\quad r_e = e^{r} - 1
\end{aligned}
$$

---


## 3. Sequence of Payments

The present value is the discounted value of all future cash flows.

$$
PV = \sum_{i=1}^{n} C_i \cdot D(t_i)
$$

where:

- $C_i$ : cash flow at time $t_i$
- $D(t_i)$ : discount factor at time $t_i$

---

### 3.1 Ordinary Annuity (Annual Compounding)

- Assume a constant payment amount $C_i = C$ for $i = 1, \dots, n$, and an annual effective interest rate $r$.
- If payments occur at the end of each year ($t_i = i$), then the discount factor is
$$
D(t_i) = (1+r)^{-i}.
$$

Therefore, the present value becomes

$$
PV = \sum_{i=1}^{n} C(1+r)^{-i}
   = C \sum_{i=1}^{n} \left(\frac{1}{1+r}\right)^i
$$

Using the Geometric Series Formula, we get:

$$
\boxed{
PV = C \frac{1-(1+r)^{-n}}{r}
}
$$

<span style="color:orange">

---

### 3.2 Geometric Series Formula

General Geometric Series Formula starting with $i=0$:

$$
\sum_{i=0}^{n} q^i = \frac{1-q^{n+1}}{1-q}
$$

Since the summation starts at $i=1$, we must use

$$
\sum_{i=1}^{n} q^i = \frac{q(1-q^n)}{1-q}
$$

where

$$
q = \frac{1}{1+r}
$$

Substituting $q = \frac{1}{1+r}$

$$
PV = C \cdot
\frac{\frac{1}{1+r}
\left(1-\left(\frac{1}{1+r}\right)^n\right)}
{1-\frac{1}{1+r}}
$$

Simplifying,

$$
PV = C \cdot \frac{1-(1+r)^{-n}}{r}
$$

## 4. Mortgage Payment

For a fixed-rate mortgage with constant periodic payments, the loan is valued as an ordinary annuity.  

$$
\text{Annuity Factor} = \frac{1-(1+r)^{-n}}{r}
$$

The annuity payment $C = \frac{PV}{\text{Annuity Factor}}$

$$
\boxed{
C = \frac{PV}{\frac{1-(1+r)^{-n}}{r}}
}
$$

---
