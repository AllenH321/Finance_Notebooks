# Option Greeks (Black–Scholes Framework)

Let:
- $V = V(t, S)$ = option price  
- $S$ = underlying price  
- $\sigma$ = volatility  
- $r$ = risk-free rate  
- $T$ = time to maturity  

Define:

$$
d_1 = \frac{\ln(S/K) + (r + \tfrac{1}{2}\sigma^2)T}{\sigma \sqrt{T}}, 
\quad
d_2 = d_1 - \sigma \sqrt{T}
$$

Let:
- $\Phi(\cdot)$ = standard normal CDF  
- $\phi(\cdot)$ = standard normal PDF  

---

## 1. Delta (Δ)

### Definition

$$
\Delta = \frac{\partial V}{\partial S}
$$

---

### Interpretation

- Sensitivity to stock price  
- Also the **hedge ratio**

---

### Formula

Call:

$$
\Delta_{call} = \Phi(d_1)
$$

Put:

$$
\Delta_{put} = \Phi(d_1) - 1
$$

---

## 2. Gamma (Γ)

### Definition

$$
\Gamma = \frac{\partial^2 V}{\partial S^2}
$$

---

### Interpretation

- Measures curvature of option price  
- Sensitivity of Delta  

---

### Formula (same for call & put)

$$
\Gamma = \frac{\phi(d_1)}{S \sigma \sqrt{T}}
$$

---

## 3. Theta (Θ)

### Definition

$$
\Theta = \frac{\partial V}{\partial t}
$$

---

### Interpretation

- Measures **time decay**
- Usually negative  

---

### Formula

Call:

$$
\Theta_{call}
=
-\frac{S \phi(d_1)\sigma}{2\sqrt{T}}
- rK e^{-rT} \Phi(d_2)
$$

Put:

$$
\Theta_{put}
=
-\frac{S \phi(d_1)\sigma}{2\sqrt{T}}
+ rK e^{-rT} \Phi(-d_2)
$$

---

## 4. Vega (ν)

### Definition

$$
\nu = \frac{\partial V}{\partial \sigma}
$$

---

### Interpretation

- Sensitivity to volatility  
- Same for call and put  

---

### Formula

$$
\nu = S \phi(d_1)\sqrt{T}
$$

---

## 5. Rho (ρ)

### Definition

$$
\rho = \frac{\partial V}{\partial r}
$$

---

### Interpretation

- Sensitivity to interest rate  

---

### Formula

Call:

$$
\rho_{call} = K T e^{-rT} \Phi(d_2)
$$

Put:

$$
\rho_{put} = -K T e^{-rT} \Phi(-d_2)
$$

---

## 🔥 Summary Table

| Greek | Definition | Meaning |
|------|----------|--------|
| Δ | $\frac{\partial V}{\partial S}$ | price sensitivity |
| Γ | $\frac{\partial^2 V}{\partial S^2}$ | curvature |
| Θ | $\frac{\partial V}{\partial t}$ | time decay |
| ν | $\frac{\partial V}{\partial \sigma}$ | volatility sensitivity |
| ρ | $\frac{\partial V}{\partial r}$ | rate sensitivity |
