# Capital Asset Pricing Model


## 1. The Riskiness of Assets

Firstly, we define the following variables:

The stock price: 
- $S(T): \Omega \to [0,\infty)$
, where $\Omega$ is the set of possible future states of the world.

The return on the risky asset over the period $[0,T]$ is
- $K = \frac{S(T)-S(0)}{S(0)}$
equivalently, 
$S(T) = S(0)(1+K)$.

The expected return of the risky asset is
- $\mu_K = E[K] = \frac{E[S(T)]-S(0)}{S(0)}.$
therefore, the expected future stock price is
$E[S(T)] = S(0)(1+\mu_K)$.

The risk of the asset is measured by the variance of the asset:
- $\sigma_K = \sqrt{\operatorname{Var}(K)}$
, and
$\sigma_K^2 = \operatorname{Var}(K) = E[K^2] - (E[K])^2$


<span style="color:blue">

### Sample: Risky assets have different returns and risk.

Suppose there are three risky assets: $S_1$, $S_2$, and $S_3$. For all three assets, there are two scenarios leading to differnt possible returns ($K_1$, $K_2$, and $K_3$) with equal probability.

| Scenario | Probability | $S_1$ & $K_1$ | $S_2$ & $K_2$ | $S_3$ $K_3$ |
|---|---:|---:|---:|---:|
| $\omega_1$ | $\frac{1}{2}$ | 14% | 13% | 25% |
| $\omega_2$ | $\frac{1}{2}$ | 14% | 15% | 3% |

For Asset 1, the return is 14% in every state. Hence,
$\mu_{K_1} = E[K_1] = \frac{1}{2}(14\%) + \frac{1}{2}(14\%) = 14\%.$

Its variance is
$\sigma_{K_1}^2 = E[K_1^2] - E[K_1]^2.$

Since $K_1$ is always 14%,
$\sigma_{K_1}^2
= \frac{1}{2}(14\%)^2 + \frac{1}{2}(14\%)^2 - (14\%)^2
= 0.$

So Asset 1 has an expected return 14% and zero risk in this two-state example.

After computing, we have the expected returns and risk (variance) for every three assets:

|  | $S_1$ & $K_1$ | $S_2$ & $K_2$ | $S_3$ $K_3$ |
|---:|---:|---:|---:|
| $\mu_K$| 14% | 14% | 14% |
| ${\sigma_K}^2$| 0 | 0.01% | 1.21% |

We can see that all three assets have the same expected return but different return variances. The variance can be interpreted as the riskiness of an asset. In this case, $S_3$ has the highest risk among all assets, while $S_1$ is considered a risk-free asset.

## 2. Portfolio of Two Risky Assets

Now consider a portfolio made of two risky assets with returns $K_1$ and $K_2$.

Let $x_1$ be the amount invested in Asset 1 and $x_2$ be the amount invested in Asset 2. The portfolio value at time 0 is

$$
V(0) = x_1 \cdot S_1(0) + x_2 \cdot S_2(0).
$$

Let $w_1$ be the weight of capital allocated in Asset 1 and $w_2$ be the weight allocated in Asset 2.
We have

$$
w_1 = \frac{x_1 \cdot S_1(0)}{V(0)}, w_2 = \frac{x_2 \cdot S_2(0)}{V(0)}
$$

where $w_1 + w_2 = 1$
- $w_i$ can be negative if we allow shorting.

The expected portfolio return and variance:

$$
\boxed{
\begin{gathered}
\mu_V = E[K_V] = w_1\mu_1 + w_2\mu_2. \\[0.8em]
\sigma_V^2 = w_1^2\sigma_1^2 + w_2^2\sigma_2^2 + 2w_1w_2\operatorname{Cov}(K_1,K_2).
\end{gathered}
}
$$

where,

$$
\operatorname{Cov}(K_1,K_2) = \rho_{12}\sigma_1\sigma_2,
$$

With correlation,

$$
\sigma_V^2
= w_1^2\sigma_1^2 + w_2^2\sigma_2^2 + 2w_1w_2\rho_{12}\sigma_1\sigma_2.
$$

---

### 2.1 Feasible Set of A Portfolio with 2 Risky Assets

Let us consider a Risk-Return Plane, where the x-axis represents the variance (risk) and the y-axis represents the expected value (return).

- Asset 1 has coordinates: $(\sigma_1, \mu_1)$, and Asset 2 has coordinates: $(\sigma_2, \mu_2)$.
- By choosing a weight $w_i$, a portfolio is created which has coordinates $(\sigma_V, \mu_V)$.
    - As the weight $w_i$ changes, the portfolio point moves in the risk-return plane.  

The collection of all possible portfolio points is called the **feasible set**. The feasible set shows every possible combination of risk and expected return that can be created from the available assets.

$$
\boxed{
\text{Feasible set} = \text{the collection of all obtainable portfolios}
}
$$

Plotting these points in the Risk-Return Plane, we obtain a line representing the feasible set of obtainable portfolios that include Assets 1 and 2. The part between the two asset points is the non-shorting zone, shown in orange.

### 3.2 Correlation and Portfolio Shape

The shape of the feasible set depends heavily on the correlation $\rho_{12}$ between the two assets.

### Case 1: Perfect Positive Correlation, $\rho_{12}=1$

If the two assets are perfectly positively correlated, then: $\rho_{12}=1$

The variance becomes:

$$
\sigma_V^2 
= w_1^2\sigma_1^2 + w_2^2\sigma_2^2 + 2w_1w_2\sigma_1\sigma_2
$$

This is a perfect square: $\sigma_V^2 = \left(w_1\sigma_1 + w_2\sigma_2\right)^2$

So:

$$
\sigma_V = w_1\sigma_1 + w_2\sigma_2
$$

In this case, both expected return and standard deviation are weighted averages of the two assets.  
Therefore, the feasible set becomes a **straight line** between the two assets in the risk-return plane.

### Case 2: Perfect Negative Correlation, $\rho_{12}=-1$

If the two assets are perfectly negatively correlated, then: $\rho_{12}=-1$

The variance becomes:

$$
\sigma_V^2 
= w_1^2\sigma_1^2 + w_2^2\sigma_2^2 - 2w_1w_2\sigma_1\sigma_2
$$

This can also be written as a perfect square: $\sigma_V^2 = \left(w_1\sigma_1 - w_2\sigma_2\right)^2$

So:

$$
\sigma_V = \left|w_1\sigma_1 - w_2\sigma_2\right|
$$

In this case, the portfolio's risk can be reduced much more effectively. If the weights are chosen correctly, it may even be possible to create a portfolio with zero risk: $w_1\sigma_1 = w_2\sigma_2$

Given $w_1 + w_2 = 1$, solving for $w_1$:

$$
w_1 = \frac{\sigma_2}{\sigma_1+\sigma_2}
$$

This is the weight that creates a risk-free portfolio when the two risky assets are perfectly negatively correlated.
