## 1. No-Arbitrage Principle

No-Arbitrage Principle says: there are no risk-free profit opportunities in a well-functioning market. If such an opportunity exists, traders will exploit it immediately, and prices will adjust until the opportunity disappears.

In other words, the probability for an admissible portfolio with initial value $V(0)=0$ such that $V(1)>0$ is **zero**.

An arbitrage opportunity has three features:
- Zero initial investment
- No risk
- Positive profit

## 2. Forwards

A **forward contract** is an agreement today (at time 0) to:
Buy or sell an asset at a fixed future date $T$ for a negotiated price (called **Future Price**) $F$.
- No money changes hands today (usually)

### Payoff Diagram

A long forward position: you agree to **buy** the asset at time $T$ for price $F$.
- Payoff for a long position at maturity: $S_t-F$

A short forward position: you agree to **sell** the asset at time $T$ for price $F$
- Payoff for a short position at maturity: $F-S_t$



### Pricing a Forward (No Dividends Model)

Under no-arbitrage, the forward price equals the spot price grown at the risk-free rate. General forward pricing formula:

$$
F(t,T)=\frac{S(t)}{B(t,T)}
$$

At time 0,

$$
\boxed{
F(0,T)=\frac{S(0)}{B(0,T)}
}
$$

With discrete annual compounding:

$$
F(0,T)=S_0(1+r)^T
$$

With continuous compounding:

$$
F(0,T)=S_0e^{rT}
$$

<span style="color:orange">
    
### Derivation Using No-arbitrage Replication

Suppose,

$S_0$: the spot price at time $T_0$

$r$: annual compounding risk-free rate

At time $T_0$:
- Borrow $S_0$ at the rate $r$ ($S_0$ inflow)
- Buy 1 share of stock at spot price $S_0$ ($S_0$ outflow)
- Short 1 forward contract at price $F$

Net cost = 0

At time $T$:
- Pay back the loan ($S_0(1+r)^T$ outflow)
- Execute the forward contract to sell one share of stock ($F$ inflow)

Net payoff $=F-S_0(1+r)^T$

Under the no-arbitrage principle, the risk-free payoff should be 0:
$$
F=S_0(1+r)^T
$$

### Arbitrage Cases

(1) Forward is overpriced
$$
F>S_0(1+r)^T
$$

At time $T_0$:
- Borrow $S_0$ at the rate $r$ ($S_0$ inflow)
- Buy 1 share of stock at spot price $S_0$ ($S_0$ outflow)
- Short 1 forward contract at price $F$

Net cost = 0

At time $T$:
- Pay back the loan ($S_0(1+r)^T$ outflow)
- Execute the forward contract to sell one share of stock ($F$ inflow)

Net payoff $=F-S_0(1+r)^T$

(2) Forward is underpriced
$$
F<S_0(1+r)^T
$$

At time $T_0$:
- Short 1 share of stock at the spot price $S_0$ ($S_0$ inflow)
- Invest $S_0$ at the risk-free rate $r$ ($S_0$ outflow)
- Long 1 forward contract at price $F$

Net cost = 0

At time $T$:
- Recieve $S_0(1+r)^T$ from the investment ($S_0(1+r)^T$ inflow)
- Execute the forward contract to buy 1 share of stock at price $F$ ($F$ outflow)
- Use the 1 share of stock to cover the short position

Net payoff $= S_0(1+r)^T - F$

</span>

### Evaluation of Forwards

The forward value equals the current spot price minus the future value of the initial forward price

With continuous compounding:
$$
\boxed{
V(t)=S(t) - S(0)e^{rt}
}
$$

<span style="color:orange">
    
Derivation:
$$
\begin{aligned}
V(t)
&= \big[ F(t,T) - F(0,T) \big] \, B(t,T) \\[6pt]
&= \left[ \frac{S(t)}{B(t,T)} - \frac{S(0)}{B(0,T)} \right] B(t,T) \\[6pt]
&= S(t) - S(0)\frac{B(t,T)}{B(0,T)} \\[6pt]
&= S(t) - S(0)e^{rt}
\end{aligned}
$$

</span>

## 3. Futures

Future contracts are agreements to buy or sell an asset at a specified price on a specified future date.

Compared to forward contracts, futures contracts have the following characteristics:
- Standardized assets, delivery date, and location
- Daily settlement (mark-to-market)

### Daily Settlement
Gains and losses are settled daily rather than only at maturity.
$$
\text{Daily Payoff}=f(t_n,T)-f(t_{n-1},T)
$$

where $f(t_n,T)$ is the future price for an asset at time $t_n$ for delivery date $T$

### Margin Account
Unlike forward contracts (OTC), futures contracts are exchange-traded and use a clearinghouse. The exchange must protect against default. So traders must post a margin — a performance bond
- Initial Margin: deposit required to open a futures position (percentage of position).
- Maintenance Margin: minimum balance required to keep the position open
- Margin Call: hit the maintenance margin, request to replenish the margin back to the initial level
