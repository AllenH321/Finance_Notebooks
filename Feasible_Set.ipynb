```python
import numpy as np
import matplotlib.pyplot as plt

# Example values
m1 = 0.10      # expected return of Asset 1
s1 = 0.10      # standard deviation of Asset 1
m2 = 0.30      # expected return of Asset 2
s2 = 0.12      # standard deviation of Asset 2
correlation = 0.5

# Portfolio weight range
# t is the weight on Asset 1
t_values = np.linspace(-3, 5, 1000)

# Covariance
c12 = correlation * s1 * s2

# Portfolio coordinates
sigma_values = []
mu_values = []

for t in t_values:
    sigma = np.sqrt(
        t**2 * s1**2 
        + (1 - t)**2 * s2**2 
        + 2 * t * (1 - t) * c12
    )
    mu = t * m1 + (1 - t) * m2
    
    sigma_values.append(sigma)
    mu_values.append(mu)

sigma_values = np.array(sigma_values)
mu_values = np.array(mu_values)

# No-shorting zone: 0 <= t <= 1
no_short = (t_values >= 0) & (t_values <= 1)

# Plot
plt.figure(figsize=(8, 6))

# Full feasible set
plt.plot(sigma_values, mu_values, linewidth=2, label="Feasible Set")

# No-shorting part
plt.plot(
    sigma_values[no_short], 
    mu_values[no_short], 
    linewidth=2, 
    label="No-Shorting Zone"
)

# Asset points
plt.scatter(s1, m1, s=20)
plt.scatter(s2, m2, s=20)

# Asset labels
plt.text(s1 + 0.01, m1, "Asset 1", fontsize=12)
plt.text(s2 + 0.01, m2, "Asset 2", fontsize=12)

# Axis labels
plt.xlabel("Risk, $\\sigma$")
plt.ylabel("Expected Return, $\\mu$")

# Title
plt.title("Feasible Set of Two Risky Assets")

# Axis limits
plt.xlim(0, 0.4)
plt.ylim(-0.1, 0.7)

# Remove background grid lines
plt.grid(False)

# Optional legend
plt.legend()

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider
```

```python
# Let's do all our work in a function, so that we can pass that function to a widget
# This will allow the widget to re-draw our graph and give us a nice interface.

def plot_feasible(correlation, m1, s1, m2, s2):
    # These are s values, my parameter
    pX = np.linspace(-3, 5, 1000)
    
    c12 = correlation*s1*s2
    
    noshortX = []
    noshortY = []
    graphX = []
    graphY = []
    for t in pX:
        # x is sigma
        x = np.sqrt(t**2 * s1**2 + (1-t)**2 * s2**2 + 2*t*(1-t)*c12)
        # y is mu
        y = t*m1 + (1-t)*m2
        
        graphX.append(x)
        graphY.append(y)
        if t >= 0 and t <= 1:
            noshortX.append(x)
            noshortY.append(y)

    plt.figure(figsize=(15,10))
    plt.plot(graphX, graphY, 'm')
    plt.plot(noshortX, noshortY, linewidth=3, color='r')
    plt.plot(s1, m1, 'bo')
    plt.plot(s2, m2, 'go')
    
    plt.title("Feasible portfolios")
    
    # Warning, these parameter are hardcoded and a bit clumsy
    plt.xlim(0, 0.4)
    plt.ylim(-0.1, 0.7)
    plt.xlabel('$\sigma$')
    plt.ylabel('$\mu$')
    
    # Plot the ratios on the graph so that we can see the criterion in Corollary 3.20
    if s2 != 0:
        risk_ratio = s1/s2
    else:
        risk_ratio = -1
    plt.text(0.3, 0.3, "$\sigma_1/\sigma_2$="+  "{:.3f}".format(risk_ratio) , fontsize=16, color="k")
    plt.text(0.3, 0.25, "correlation=" + "{:.3f}".format(correlation) , fontsize=16, color="k")

interact(
    plot_feasible,
    correlation=FloatSlider(min=-1.0, max=1.0, step=0.05, value=0.5),
    m1=FloatSlider(min=0, max=1.0, step=0.05, value=0.1),
    s1=FloatSlider(min=0, max=0.5, step=0.001, value=0.1),
    m2=FloatSlider(min=0, max=1.0, step=0.05, value=0.3),
    s2=FloatSlider(min=0, max=0.5, step=0.001, value=0.12)
)
```
