# Random Walk & Brownian Motion

- Random Walk

## 1. Symmetrical Random Walk (One-Step Increment)

Let $X_i$ be the increment at step $i$:

$$
X_i =
\begin{cases}
+1 & \text{with probability } \frac{1}{2} \\
-1 & \text{with probability } \frac{1}{2}
\end{cases}
$$

### Properties

- Expected value:
  $$
  \mathbb{E}[X_i] = 0
  $$

- Variance:
  $$
  \mathrm{Var}(X_i) = 1
  $$

## 1.1 Random Walk Process

Define the random walk $M_k$ as the cumulative sum of increments:

$$
M_0 = 0
$$

$$
M_k = \sum_{i=1}^{k} X_i
$$

- Expected value:
  $$
  \mathbb{E}[M_k] = 0
  $$

- Variance:
  $$
  \mathrm{Var}(M_k) = k
  $$

Each step adds a random increment $X_i$ to the random walk process. The process evolves as a sum of independent random variables
- Over time:
  - Mean remains constant (no drift)
  - Variance grows linearly with time

## 1.2 Scaled Symmetric Random Walk

Take a symmetric random walk and shrink both time and step size so it becomes a continuous process.

Consider a symmetric random walk,

- $X_i = \pm 1$ with probaility $\frac{1}{2}$
- $M_k = \sum^{k}_{i=1}X_i$

This is a discrete process with jumps of size 1 and time steps of size 1

We chop each unit of time into more steps and smaller sizes:

Given $n$ steps per unit time, the walk takes steps of size $\frac{1}{\sqrt n}$ at intervals of $\frac{1}{n}$

$$
W^{(n)}(t)=\frac{1}{\sqrt{n}}M_{nt}
$$

- Expected Value:
$$
E[W^{(n)}(t)]=0
$$

- Variance:
$$
Var(W^{(n)}(t))=\frac{1}{n} \cdot nt = t
$$

**From Scaled Symmetric Random Walk to Brownian Motion**

According to the Central Limit Theorem,

Let $X_1$, $X_2$,...,$X_n$ be i.i.d. with:

$$
E[X_i] = \mu, Var(X_i)=\sigma^2
$$

Then,

$$
\sqrt{n}(\bar{X} - \mu) \xrightarrow{d} \mathcal{N}(0, \sigma^2)
$$

where, 

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

or,

$$
\boxed{
\frac{1}{\sqrt n} \sum_{i=1}^{n} (X_1 - \mu) \xrightarrow{d} \mathcal{N}(0,\sigma^2)
}
$$

Now, recall the scaled symmetric random walk:

$$
W^{(n)}(t) = \frac{1}{\sqrt{n}} M_{nt}
$$

Rewrite the steps,

$$
W^{(n)}(t) = \frac{1}{\sqrt{n}} \sum_{j=1}^{nt} X_j
$$

Now factor it like CLT:

$$
= \frac{1}{\sqrt{nt}} \sum_{j=1}^{nt} X_j \cdot \sqrt{t}
$$

So:

$$
W^{(n)}(t) = \sqrt{t} \cdot \frac{1}{\sqrt{nt}} \sum_{j=1}^{nt} X_j
$$

By CLT:

$$
W^{(n)}(t) \xrightarrow{d} \sqrt{t} \cdot \mathcal{N}(0,1)
$$

Simplify:

$$
\boxed{
W^{(n)}(t) \xrightarrow{d} \mathcal{N}(0,t)
}
$$

This matches the distribution of Brownian Motion.

## 3. Brownian Motion (Wiener Process)

Let $(\Omega, \mathcal{F}, \mathbb{P})$ be a probability space.

For every $\omega \in \Omega$, there exists a continuous function $W(t)$ for $t \ge 0$, with $W(0) = 0$ and $W(t)$ depends on the path $\omega$.

We say that $W(t)$ is a **Brownian Motion (Wiener Process)** if, for all

$$
0 = t_0 < t_1 < t_2 < \cdots < t_n,
$$

the following properties hold:

### Independent Increments

$$
W(t_{i+1}) - W(t_i) \text{ are independent}
$$

### Mean and Variance

$$
\mathbb{E}[W(t_{i+1}) - W(t_i)] = 0
$$

$$
\mathrm{Var}(W(t_{i+1}) - W(t_i)) = t_{i+1} - t_i
$$

### Normal Distribution of Increments

$$
W(t_{i+1}) - W(t_i) \sim \mathcal{N}(0, t_{i+1} - t_i)
$$

### Markov & Martingale

**Markov property**

A stochastic process $X_t$ is Markov if:

$$
\mathbb{P}(X_t \le x \mid \mathcal{F}_s)
=\mathbb{P}(X_t \le x \mid X_s), \quad \text{for } s < t
$$

where,
- $\mathcal{F}_s$: all information up to time $s$
$X_s$: current value

The property means that the future depends only on the current state, not the past.

**Martingale Property**

A stochastic process $X_t$ is a **martingale** if:

$$
\mathbb{E}[X_t \mid \mathcal{F}_s] = X_s, \quad \text{for } s < t
$$

The property means that the best prediction of the future value is the current value.

### Proof: Symmetric Random Walk is a martingale

For a symmetric random walk $M_n$, 

$$
\mathbb{E}[M_\ell \mid \mathcal{F}_k] = M_k, \text{for } \ell > k
$$

<span style="color:orange">

Decompose $M_\ell$

$$
M_\ell = (M_\ell - M_k) + M_k
$$

So,

$$
\mathbb{E}[M_\ell \mid \mathcal{F}_k]
= \mathbb{E}[M_\ell - M_k + M_k \mid \mathcal{F}_k]
$$

Linearity of conditional expectation

$$
\mathbb{E}[M_\ell \mid \mathcal{F}_k]
= \mathbb{E}[M_\ell - M_k \mid \mathcal{F}_k]+ \mathbb{E}[M_k \mid \mathcal{F}_k]
$$

Second term $M_k$ is $\mathcal{F}_k$-measurable,

$\mathbb{E}[M_k \mid \mathcal{F}_k] = M_k$

First term $M_\ell - M_k = \sum_{i=k+1}^{\ell} X_i$ is the future increment,
- independent of $\mathcal{F}_k$
- the expected value of the increment is $0$

So,

$$
\mathbb{E}[M_\ell - M_k \mid \mathcal{F}_k]
= \mathbb{E}[M_\ell - M_k]
= 0
$$

and,

$$
\mathbb{E}[M_\ell \mid \mathcal{F}_k]
= 0 + M_k
= M_k
$$

### Proof: Brownian Motion is a Martingale

Let $(W(t))_{t \ge 0}$ be a Brownian motion defined on a probability space 
$(\Omega, \mathcal{F}, \mathbb{P})$ with natural filtration $\mathcal{F}_s$.

We want to show that for $0 \le s < t$,

$$
\mathbb{E}[W(t) \mid \mathcal{F}_s] = W(s)
$$

Decompose $W(t)$,

$$
W(t) = (W(t) - W(s)) + W(s)
$$

So,

$$
\mathbb{E}[W(t) \mid \mathcal{F}_s]
= \mathbb{E}[W(t) - W(s) + W(s) \mid \mathcal{F}_s]
$$

Linearity of conditional expectation

$$
= \mathbb{E}[W(t) - W(s) \mid \mathcal{F}_s]
+ \mathbb{E}[W(s) \mid \mathcal{F}_s]
$$

Second term $W(s)$ is $\mathcal{F}_s$-measurable,

$$
\mathbb{E}[W(s) \mid \mathcal{F}_s] = W(s)
$$

First term $W(t) - W(s)$ is the future increment, which:

- is independent of $\mathcal{F}_s$
- has mean $0$
- satisfies $W(t) - W(s) \sim \mathcal{N}(0, t - s)$

So,

$$
\mathbb{E}[W(t) - W(s) \mid \mathcal{F}_s]
= \mathbb{E}[W(t) - W(s)]
= 0
$$

In conclusion,

$$
\mathbb{E}[W(t) \mid \mathcal{F}_s]
= 0 + W(s)
= W(s)
$$
