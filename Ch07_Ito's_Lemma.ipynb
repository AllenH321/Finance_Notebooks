## Quadratic Variation

Take a continuous function $f(t)$ on $[0,T]$.

- Partition the interval:

$$
0 = t_0 < t_1 < \cdots < t_n = T
$$

- Define mesh:

$$
\|P\| = \max (t_{i+1} - t_i)
$$

Then the quadratic variation is:

$$
[f,f](T) = \lim_{\|P\| \to 0} \sum_{i=0}^{n-1} \left(f(t_{i+1}) - f(t_i)\right)^2
$$

Quadratic variation measures the accumulated squared increments of a process and captures its path roughness.

### Smooth function (e.g., $f(t) = t^2$)

- increments $\approx$ small  
- squared increments $\approx$ very small  

$$
[f,f](T) = 0
$$

### Brownian motion

- increments $\approx \sqrt{\Delta t}$  
- squared increments $\approx \Delta t$

Sum:

$$
\sum \Delta t = T
$$

So:

$$
[W,W](T) = T
$$

## Stochastic Calculus

Properties of stochastic calculus:
- $(dt)^2 = 0$
- $dWdt = 0$
- **$(dW)^2 = dt$**, from quadratic variation $[W,W](T)=T$

Consider $f(t, W(t))$ and $f \in C^2$ (twice continuously differentiable),

So $f$ depends on:
- time $t$
- the Wiener process $W(t)$

Using multivariable Taylor expansion:

$$
df = f_t dt + f_x dW + \frac{1}{2} f_{xx} (dW)^2 + ...
$$

From the properties of stochastic calculus,

$$
(dW)^2 = dt
$$

In conclusion,

$$
df = f_t dt + f_x dW + \frac{1}{2} f_{xx} dt
$$

## Stochastic Differential Equation (SDE)

For a very small time interval $\Delta t$, assume the stock's **relative change** is:

$$
\frac{\Delta S}{S} \approx \mu \Delta t + \sigma \Delta W
$$

or equivalently,

$$
\frac{S_{t+\Delta t} - S_t}{S_t} \approx \mu \Delta t + \sigma \Delta W_t
$$

where:

- $\mu$ = drift (average rate of return)  
- $\sigma$ = volatility  
- $\Delta W_t$ = random shock over the interval $\Delta t$

For Brownian motion,

$$
\Delta W_t \sim \mathcal{N}(0, \Delta t)
$$

so the random shock has size about $\sqrt{\Delta t}$.

Move to Continuous Time:

Now let $\Delta t \to 0$. In differential notation:

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

## Ito's Formula

Let $X_t = W_t$ and take a smooth function $f(x)$.

Use Taylor expansion:

$$
df= f'(W_t)dW_t + \frac{1}{2}f''(W_t)(dW_t)^2+...
$$

## Itô Integral

The Itô integral is:

$$
\int_0^T f(t, W(t))\, dW(t)
$$

Why we need a sepcial definition:
- $W(t)$ is not differentiable  
- It is extremely rough (infinite variation)  

So $dW(t)$ is not a normal “small change.”


Itô defines it as a limit of sums:

$$
\int_0^T f(t, W(t))\, dW(t)
=\lim_{\|P\|\to 0}
\sum_{i=0}^{n-1}
f(t_i, W(t_i)) \cdot \big(W(t_{i+1}) - W(t_i)\big)
$$

The function is evaluated at the **left endpoint**:

$$
f(t_i, W(t_i))
$$

This is called: **non-anticipative / adapted**
- you only use information available up to time $t_i$  
- you don’t “look into the future”

## Interpretation

It represents the accumulation of random shocks over time

Each term: $f(t_i, W(t_i)) \cdot (W(t_{i+1} - W_{t_i})$

Itô integrals are martingales:

$$
\mathbb{E}\left[\int_0^T f(t, W(t))\, dW(t)\right] = 0
$$

## Finance interpretation

In finance:

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

The solution:

$$
S_t = S_0 + \int_0^t \mu S_u\,du + \int_0^t \sigma S_u\,dW_u
$$


## Solving the Stock Price SDE Using Itô’s Formula


The stock price follows the SDE:

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

transformation

Let $f(S_t) = \ln S_t$

Then:

$$
f'(S_t) = \frac{1}{S_t}, \quad f''(S_t) = -\frac{1}{S_t^2}
$$

Apply Itô’s formula

$$
d(\ln S_t)
= f'(S_t)dS_t + \frac{1}{2} f''(S_t)(dS_t)^2
$$

So:

$$
d(\ln S_t)
= \frac{1}{S_t} dS_t - \frac{1}{2}\frac{1}{S_t^2}(dS_t)^2
$$

Substitute the SDE

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

First term: $\frac{1}{S_t} dS_t = \mu dt + \sigma dW_t$

Compute $(dS_t)^2$

$$
(dS_t)^2 = (\sigma S_t dW_t)^2 = \sigma^2 S_t^2 (dW_t)^2
$$

Using:

$$
(dW_t)^2 = dt
$$

we get:

$$
(dS_t)^2 = \sigma^2 S_t^2 dt
$$

Plug back

$$
-\frac{1}{2}\frac{1}{S_t^2}(dS_t)^2
= -\frac{1}{2}\sigma^2 dt
$$

Combine terms

$$
d(\ln S_t)
= \mu dt + \sigma dW_t - \frac{1}{2}\sigma^2 dt
$$

$$
= \left(\mu - \frac{1}{2}\sigma^2\right)dt + \sigma dW_t
$$

Integrate

$$
\ln S(t) - \ln S(0)
= \left(\mu - \frac{1}{2}\sigma^2\right)t + \sigma W(t)
$$

Exponentiate

$$
S(t)
= S(0)\exp\Big((\mu - \tfrac{1}{2}\sigma^2)t + \sigma W(t)\Big)
$$

Final Result

$$
\boxed{
S(t) = S(0)\exp\Big((\mu - \tfrac{1}{2}\sigma^2)t + \sigma W(t)\Big)
}
$$

## Verifying the Solution of the Stock Price SDE Using Itô’s Formula

We start with a candidate solution of the form:

$$
S(t, x) = S(0)\exp\Big((v - \tfrac{1}{2}\sigma^2)t + \sigma x\Big)
$$

We will verify that this function satisfies a stochastic differential equation when we substitute $x = W(t)$.

---

### Step 1: Compute partial derivatives

First, compute the derivatives of $S(t,x)$:

$$
\frac{\partial S}{\partial t}
= S(t,x)\Big(v - \tfrac{1}{2}\sigma^2\Big)
$$

$$
\frac{\partial S}{\partial x}
= \sigma S(t,x)
$$

$$
\frac{\partial^2 S}{\partial x^2}
= \sigma^2 S(t,x)
$$

---

### Step 2: Apply Itô’s formula

Using Itô’s formula:

$$
dS = S_t\,dt + S_x\,dW(t) + \frac{1}{2} S_{xx}\,dt
$$

Substitute the derivatives:

$$
dS
= S\Big(v - \tfrac{1}{2}\sigma^2\Big)dt
+ \sigma S\,dW(t)
+ \frac{1}{2}\sigma^2 S\,dt
$$

---

### Step 3: Simplify

Combine the $dt$ terms:

$$
dS
= v S\,dt - \tfrac{1}{2}\sigma^2 S\,dt + \sigma S\,dW(t) + \tfrac{1}{2}\sigma^2 S\,dt
$$

The correction terms cancel:

$$
dS = v S\,dt + \sigma S\,dW(t)
$$

---

### Step 4: Express in relative form

$$
\frac{dS}{S} = v\,dt + \sigma\,dW(t)
$$

---

### Step 5: Interpretation

We have shown that the exponential function satisfies the SDE:

$$
dS = v S\,dt + \sigma S\,dW(t)
$$

To match the standard stock price model:

$$
dS = \mu S\,dt + \sigma S\,dW(t)
$$

we identify:

$$
v = \mu + \tfrac{1}{2}\sigma^2
$$

---

### Step 6: Risk-neutral interpretation

Under the risk-neutral measure:

$$
\mu \rightarrow r
$$

so:

$$
v = r + \tfrac{1}{2}\sigma^2
$$

---

## Final Insight

We verified that:

$$
S(t) = S(0)\exp\Big((v - \tfrac{1}{2}\sigma^2)t + \sigma W(t)\Big)
$$

is the solution of the SDE

$$
dS = v S\,dt + \sigma S\,dW(t)
$$

This confirms that the exponential form correctly models geometric Brownian motion.
