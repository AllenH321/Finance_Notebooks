## 1. Options

An option is a contract that gives the holder the **right, but not the obligation**, to buy or sell an asset at a fixed price (called **strike price**) $K$ before or at maturity $T$.
- American options can be exercised before maturity.
- European options can only be exercised at maturity.

### Payoff

A long call gives you the right to buy the asset for $K$.
- Payoff of a long call: $(S_T-K)^{+}$ (Graph 1)
- Premium paid
  
A short call means you sell a call (others can buy the asset from you) for $K$.
- Payoff of a short call: $-(S_T-K)^{+}$ (Graph 3)
- Premium received
  
A long put gives you the right to sell the asset for $K$.
- Payoff of a long put: $(K-S_T)^{+}$ (Graph 2)
- Premium paid
  
A long put means you sell a put (others can sell the asset to you) for $K$.
- Payoff of a short put: $-(K-S_T)^{+}$ (Graph 4)
- Premium received



## 2. Baskets of Options

(1) Straddle (Long Straddle)
- Long 1 European Call at strike $K$
- Short 1 European Put at strike $K$

Profit from large price movements in either direction.

(2) Spread (Call Spread)
- Long 1 European Call at strike $K_1$
- Short 1 European Call at strike $K_2$


(3) Strangle (Long Strangle)
- Long 1 European Put at strike $K_1$
- Long 1 European Call at strike $K_2$, where $K_1<K_2$

(4) Bufferfly
- Long 1 European Call at strike $K_1$
- Short 2 European Call at strike $K_2$, where $K_2 = K_1 + \lambda$
- Long 1 European Call at strike $K_3$, where $K_3 = K_2 + \lambda$

(5) Iron Condor
- Long 1 European Call at strike $K_1$
- Short 1 European Call at strike $K_2$
- Short 1 European Call at strike $K_3$
- Long 1 European Call at strike $K_4$, where $K_1<K_2<K_3<K_4$

## 3. Put-Call Parity

Put-Call Parity states that holding a portfolio of a long call option and a short put option (or vice versa) should yield the same return as having one share of the underlying stock, assuming the same underlying asset, strike price, and expiration date.

The Put-Call Parity Formula (continuous compounding):
$$
\boxed{
C_0-P_0 = S_0-Ke^{-rT}
}
$$

<span style="color:orange">

### Derivation using payoffs:

At maturity $T_0$:

A long call payoff $C_T=(S_T-K)^+$

A short put payoff $P_T=-(K-S_T)^+$

Compute $C_T-P_T$, 

If $S_T>K$, $C_T-P_T=(S_T-K)-0=S_T-K$

If $S_T<K$, $C_T-P_T=0-(K-S_T)=S_T-K$

In all cases, 
$$
$C_T-P_T=S_T-K$
$$

Discounted to present values:
$$
C_0-P_0=S_0-Ke^{-rT}
$$

### Arbitrage Cases

(1) $C_0-P_0>S_0-Ke^{-rT}$

At time $T_0$,
- Long 1 share of the stock ($S_0$ outflow)
- Short 1 zero-coupon bond with face value $K$ ($Ke^{-rT}$ inflow)
- Short 1 European Call ($C_0$ inflow)
- Long 1 European Put ($P_0$ outflow)

Net cash flow $=-S_0+Ke^{-rT}+C_0-P_0$ = $(C_0-P_0)-(S_0-Ke^{-rT}) > 0$

At time $T,

- If $S_T>K$: Sell the stock, Pay the bond, Exercise Call, Do not exercise Put 

Net payoff $= S_T-K-(S_T-K)+0 = 0$

- If $S_T<K$: Sell the stock, Pay the bond, Do not Exercise Call, Exercise Put 

Net payoff $= S_T-K-0+(K-S_T) = 0$

(2) $C_0-P_0<S_0-Ke^{-rT}$

At time $T_0$,
- Short 1 share of the stock ($S_0$ inflow)
- Long 1 zero-coupon bond with face value $K$ ($Ke^{-rT}$ outflow)
- Long 1 European Call ($C_0$ outflow)
- Short 1 European Put ($P_0$ inflow)

Net cash flow $=S_0-Ke^{-rT}-C_0+P_0 = (S_0-Ke^{-rT})-(C_0-P_0) > 0$

At time $T,

- If $S_T>K$: Buy back the stock, Collect the bond, Exercise Call, Do not exercise Put 

Net payoff $= -S_T+K+(S_T-K)+0 = 0$

- If $S_T<K$: Buy back the stock, Collect the bond, Do not Exercise Call, Exercise Put 

Net payoff $= -S_T+K-0+(K-S_T) = 0$
