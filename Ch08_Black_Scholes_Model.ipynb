## Deriving the Black–Scholes PDE (No-Arbitrage / Perfect Hedge Approach)

### Step 0: Stock price model

Assume the stock follows:

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

---

### Step 1: Option price

Let the option price be:

$$
f = f(t, S)
$$

---

### Step 2: Apply Itô’s formula

$$
df = f_t\,dt + f_S\,dS + \frac{1}{2} f_{SS} (dS)^2
$$

Substitute:

$$
dS = \mu S\,dt + \sigma S\,dW
$$

and use $(dW)^2 = dt$:

$$
df =
\Big(f_t + \mu S f_S + \tfrac{1}{2}\sigma^2 S^2 f_{SS}\Big)dt
+ \sigma S f_S\,dW
$$

---

### Step 3: Construct a riskless portfolio

Form the portfolio:

$$
\Pi = f - \Delta S
$$

---

### Step 4: Portfolio dynamics

$$
d\Pi = df - \Delta\,dS
$$

Substitute:

$$
d\Pi =
\Big(f_t + \mu S f_S + \tfrac{1}{2}\sigma^2 S^2 f_{SS}\Big)dt
+ \sigma S f_S dW
- \Delta(\mu S dt + \sigma S dW)
$$

---

### Step 5: Eliminate randomness (perfect hedge)

Choose:

$$
\Delta = f_S
$$

Then the $dW$ terms cancel:

$$
d\Pi =
\Big(f_t + \tfrac{1}{2}\sigma^2 S^2 f_{SS}\Big)dt
$$

👉 The portfolio is now **risk-free**

---

### Step 6: Apply no-arbitrage condition

A risk-free portfolio must grow at rate $r$:

$$
d\Pi = r \Pi dt
$$

Substitute $\Pi = f - S f_S$:

$$
d\Pi = r(f - S f_S)dt
$$

---

### Step 7: Derive the PDE

Equate both expressions:

$$
f_t + \tfrac{1}{2}\sigma^2 S^2 f_{SS}
= r(f - S f_S)
$$

Rearrange:

$$
\boxed{
f_t + r S f_S + \tfrac{1}{2}\sigma^2 S^2 f_{SS} - r f = 0
}
$$

# Black–Scholes Framework: From Portfolio to Pricing

---

## 1. Stock Price Model (Real-World Measure)

We assume the stock follows a geometric Brownian motion:

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

- $\mu$: expected return (drift)
- $\sigma$: volatility
- $W_t$: Brownian motion

---

## 2. Self-Financing Portfolio

Construct a portfolio consisting of:

- $x(t)$ units of stock
- $y(t)$ units of risk-free asset

$$
V(t) = x(t) S(t) + y(t) A(t)
$$

where:

$$
dA_t = r A_t\,dt
$$

---

### Self-financing condition

$$
dV_t = x(t)\,dS_t + y(t)\,dA_t
$$

👉 No external cash injection.

---

## 3. Portfolio Dynamics (Under $P$)

Substitute:

$$
dS_t = \mu S_t dt + \sigma S_t dW_t
$$

$$
dV_t = x(\mu S dt + \sigma S dW) + y(rA dt)
$$

Rewriting:

$$
dV_t = rV_t dt + x(\mu - r)S_t dt + \sigma x S_t dW_t
$$

---

### Interpretation

- $rV_t dt$ → risk-free growth  
- $x(\mu - r)S_t dt$ → **risk premium**  
- $\sigma x S_t dW_t$ → randomness  

---

## 4. Discounting

Define discounted processes:

$$
\tilde{S}_t = e^{-rt} S_t, \quad \tilde{V}_t = e^{-rt} V_t
$$

---

### Dynamics of discounted stock

Using Itô:

$$
d\tilde{S}_t = (\mu - r)\tilde{S}_t dt + \sigma \tilde{S}_t dW_t
$$

---

## 5. Change to Risk-Neutral Measure ($Q$)

We change measure so that:

$$
d\tilde{S}_t = \sigma \tilde{S}_t dW_t^Q
$$

👉 Drift disappears.

---

### Key idea

> Under $Q$, discounted asset prices are **martingales**.

---

## 6. Discounted Portfolio Becomes a Martingale

From earlier:

$$
d\tilde{V}_t = x(t)\,d\tilde{S}_t
$$

So:

$$
d\tilde{V}_t = \sigma x(t)\tilde{S}_t dW_t^Q
$$

👉 Pure Itô integral ⇒

$$
\mathbb{E}^Q[d\tilde{V}_t] = 0
$$

---

### Martingale property

$$
\tilde{V}_0 = \mathbb{E}^Q[\tilde{V}_T]
$$

Thus:

$$
V_0 = e^{-rT}\mathbb{E}^Q[V_T]
$$

---

## 7. Apply to Option Pricing

Let payoff:

$$
V_T = g(S_T)
$$

Then:

$$
\boxed{
V_0 = e^{-rT} \mathbb{E}^Q[g(S_T)]
}
$$

👉 This is the **risk-neutral pricing formula**

---

## 8. PDE Formulation (Local Version)

Let option price:

$$
u(t,x) = \text{price at time } t \text{ when } S_t = x
$$

Apply Itô’s formula:

$$
du = u_t dt + u_x dS + \tfrac{1}{2}u_{xx} dS^2
$$

Substitute:

$$
dS = rS dt + \sigma S dW
$$

---

### Result

$$
du =
\Big(u_t + r x u_x + \tfrac{1}{2}\sigma^2 x^2 u_{xx}\Big)dt
+ \sigma x u_x dW
$$

---

## 9. Martingale Condition

Discount:

$$
e^{-rt}u(t,S_t)
$$

must be a martingale ⇒ drift = 0

---

### This gives the Black–Scholes PDE

$$
\boxed{
u_t + r x u_x + \tfrac{1}{2}\sigma^2 x^2 u_{xx} - r u = 0
}
$$

with terminal condition:

$$
u(T,x) = g(x)
$$

---

## 10. Final Connection

Two equivalent representations:

### (1) PDE
Solve:

$$
u_t + r x u_x + \tfrac{1}{2}\sigma^2 x^2 u_{xx} = r u
$$

### (2) Expectation

$$
\boxed{
u(0,S_0) = e^{-rT}\mathbb{E}^Q[g(S_T)]
}
$$

---

## 🔥 Key Insight

All three viewpoints are equivalent:

- Replicating portfolio  
- No-arbitrage (perfect hedge)  
- Risk-neutral martingale  

👉 They all lead to the same pricing rule.

---

## 🧠 One-line Summary

> Remove risk → force return = $r$ → obtain PDE → equivalent to risk-neutral expectation.

# Black–Scholes Formula: Closed-Form Derivation

---

## 1. Starting Point: Risk-Neutral Pricing

From no-arbitrage and martingale arguments:

$$
C(0) = e^{-rT} \mathbb{E}^Q[(S_T - K)^+]
$$

👉 We now compute this expectation explicitly.

---

## 2. Rewrite the Payoff

Use the indicator function:

$$
(S_T - K)^+ = \mathbf{1}_{\{S_T > K\}} (S_T - K)
$$

So:

$$
C(0)
= e^{-rT} \mathbb{E}^Q[\mathbf{1}_{\{S_T > K\}} (S_T - K)]
$$

---

## 3. Split into Two Terms

$$
C(0)
= e^{-rT}
\left(
\mathbb{E}^Q[\mathbf{1}_{\{S_T > K\}} S_T]
- K \mathbb{E}^Q[\mathbf{1}_{\{S_T > K\}}]
\right)
$$

---

### 🔑 Interpretation

- First term: **expected stock value when in-the-money**
- Second term: **probability of exercise × strike**

---

## 4. Distribution of $S_T$ under $Q$

Under the risk-neutral measure:

$$
S_T = S_0 \exp\left((r - \tfrac{1}{2}\sigma^2)T + \sigma W_T\right)
$$

where:

$$
W_T \sim N(0, T)
$$

---

### Standardization

Define:

$$
Z = \frac{W_T}{\sqrt{T}} \sim N(0,1)
$$

---

## 5. Convert Event $S_T > K$

Take logs:

$$
S_T > K
\;\Longleftrightarrow\;
\ln S_T > \ln K
$$

Substitute:

$$
\ln S_T =
\ln S_0 + (r - \tfrac{1}{2}\sigma^2)T + \sigma W_T
$$

---

### Rearranging

$$
Z > d_2
$$

where:

$$
d_2 =
\frac{\ln(S_0/K) + (r - \tfrac{1}{2}\sigma^2)T}{\sigma \sqrt{T}}
$$

---

## 6. Compute Probability Term

$$
\mathbb{E}^Q[\mathbf{1}_{\{S_T > K\}}]
= \mathbb{P}^Q(S_T > K)
= \Phi(d_2)
$$

---

## 7. Compute Stock Term

We compute:

$$
\mathbb{E}^Q[\mathbf{1}_{\{S_T > K\}} S_T]
$$

Using lognormal density and completing the square (as in your board work), we obtain:

$$
\mathbb{E}^Q[\mathbf{1}_{\{S_T > K\}} S_T]
= S_0 e^{rT} \Phi(d_1)
$$

where:

$$
d_1 =
\frac{\ln(S_0/K) + (r + \tfrac{1}{2}\sigma^2)T}{\sigma \sqrt{T}}
$$

---

### 🔑 Key relation

$$
d_1 = d_2 + \sigma \sqrt{T}
$$

---

## 8. Combine Results

Substitute into pricing formula:

$$
C(0)
= e^{-rT}
\left(
S_0 e^{rT} \Phi(d_1)
- K \Phi(d_2)
\right)
$$

---

## 9. Final Black–Scholes Formula

$$
\boxed{
C_0 = S_0 \Phi(d_1) - K e^{-rT} \Phi(d_2)
}
$$

---

## 🔥 Interpretation

- $S_0 \Phi(d_1)$ → expected stock payoff (adjusted)
- $K e^{-rT} \Phi(d_2)$ → discounted expected strike payment

---

The formula decomposes the call into:

```text
Call price =
(probability-weighted stock value)
− (probability-weighted strike payment)
