## 1. Zero-Coupon Bonds

A Zero-Coupon Bond has only a single payment. The issuing institution (for example, a government, a bank, or a company) promises to exchange the bond for a certain amount of money, $F$, called the **Face Value**, on a given time $T$, called the **Maturity Date**.

The future value of the bond is the face value $F$.

The present value of the zero-coupon bond should be:
$$
\boxed{
V(0) = F\cdot B(0,T)
}
$$

With annual compounding:
$$
V(0) = F(1+r)^{-T}
$$

---

### 1.1 Spot Rate

On the contrary, bonds are freely traded in the market, so that market forces determine their prices. The spot rate for maturity $T$ is the annualized return implied by the price of a zero-coupon bond maturing at time $T$.

The implied effective annual rate with annual compounding:

$$
r = \left(\frac{FV}{PV}\right)^{\frac{1}{T}} - 1
$$

Discrete compounding (m times per year):

$$
r = \left(\frac{FV}{PV}\right)^{\frac{1}{mT}} - 1
$$

Continuous compounding:

$$
r = \frac{1}{T}\ln\left({\frac{FV}{PV}}\right)
$$

---

## 2. Unit Bond

A Unit Bond has a face value equal to one unit of the money currency, $F=1$.

- $B(t, T)\;\; t \in [t, T]$ is the price of a unit bond with maturity $T$ at time $t$, usually used as the **Discount Factor** for the bond at time $t$.
- $B(0, T)$ is the current, time 0, price of the bond
- $B(T, T) = 1$ is equal to the face value.

The present value of the unit bond at time $t$ with annual compounding:

$$
B(t, T) = (1+r)^{-(T-t)}
$$

Discrete compounding (m times per year):

$$
B(t, T) = \left(1+\frac{r}{m}\right)^{-m(T-t)}
$$

Continuous compounding:

$$
B(t, T) = e^{-r(T-t)}
$$

---

## 3. Coupon Bond

Bonds consist of payments of the face value $F$ at maturity, and coupon payments $C$ paid regularly are called **Coupon Bonds**.

Given the annualized coupon rate $c$, the annual coupon amount $C = cF$.

The evaluation of a coupon bond is similar to the annuity payment. The present value is the discounted value of all future cash flows.

$$
PV = \sum_{i=1}^{n} C_i \cdot D(t_i)
$$

The periodic coupon payments $C$ are constant, and we have a face value $F$ paid at maturity.

$$
\boxed{
PV = \sum_{i=1}^{n} \frac{C}{m} \, B(0,t_i) + F \, B(0,T)
}
$$

The present value of the coupon bond with discrete compounding:

$$
PV = \sum_{i=1}^{n} 
\frac{\frac{C}{m}}{(1+\frac{r}{m})^i} + 
\frac{F}{(1+\frac{r}{m})^n}
$$

## 4. Clean Price, Dirty Price, and Accrued Interest

- Dirty Price (Full Price): the present value of all future payments
    - It is the price you actually pay to buy the bond.
- Accrued Interest: the interest accrued since the last payment
    - Accrued fraction $w = \frac{
    \text{days since last payment}}{\text{days in a whole payment period}}$
    - Accrued interest = $C \cdot w$
- Clean Price: The real bond price excluding the accrued interest
    - Clean Price = Dirty Price - Accrued Interest
    - The market quotes the CLEAN price.
 
### 4.1 Evaluation of Coupon Bonds between Coupon Payments

When the settlement happens between two coupon payments, all future coupon cash flows are closer by the fraction $w$.

The remaining time to each payment changes from $i = 1, 2, 3,  \dots, n$ to $i = 1-w, 2-w, 3-w, \dots, n-w$

Therefore, the present value of a coupon bond with discrete compounding:

$$
PV = \sum_{i=1}^{n} 
\frac{\frac{C}{m}}{(1+\frac{r}{m})^{i-w}} + 
\frac{F}{(1+\frac{r}{m})^{n-w}}
$$

Especially, the present value of a **unit bond** with discrete compounding:

$$
PV = \sum_{i=1}^{n} 
\frac{\frac{C}{m}}{(1+\frac{r}{m})^{i-w}} + 
\frac{1}{(1+\frac{r}{m})^{n-w}}
$$

Simplifying it, we get:
$$
\boxed{
PV = \left(1+\frac{r}{m}\right)^w
\left[\frac{C}{r}\left(1-\frac{1}{(1+\frac{r}{m})^n}\right) + 
\frac{1}{(1+\frac{r}{m})^n}\right]
}
$$

## 5. Premium Bond & Discount Bond

A **par bond** trades at its face value: $PV = F$

This occurs when the coupon rate equals the yield: $c=r$

The bond pays exactly the market-required return.

A **premium bond** trades above its face value: $PV>F$

This occurs when: $c>r$

The bond pays coupons higher than the market yield, so investors are willing to pay more than par.


A **discount bond** trades below its face value: $PV<F$

This occurs when: $c<r$

The bond pays coupons lower than the market yield, so its price must fall below par.

### The Pull-to-Par Effect

As maturity approaches:

$$
PV \to F
$$

Premium bonds decrease toward par,  
Discount bonds increase toward par.

## 6. Duration

Duration measures the sensitivity of the bond price to changes in the interest rate. We find duration by computing the first derivative of the bond price with respect to the interest rate $y$ (or $r$).

### Dollar Duration (DD)

$-\frac{\partial P}{\partial y}$ is the **Dollar Duration** (before dividing by price).
- It measures **How much** the bond price changes when the yield changes (measured in decimals).

Using first-order approximation (Taylor expansion):

$$
dP \approx \frac{dP}{dy}dy
$$

Substituting $\frac{dP}{dy}=-DD$,

$$
dP \approx -DD \cdot dy
$$

- If yield increases: 
$dy>0 \; \to \; dP<0$, meaning the bond price falls.

- If yield decreases:
$dy<0 \; \to \; dP>0$, meaning the bond price rises.


For a unit bond with discrete compounding, the dollar duration (DD) is calculated as:

$$
\boxed{
-\frac{\partial P}{\partial y} = 
\frac{C}{y^2}\left[(1+\frac{y}{m})^{-n}-1\right] +
\frac{n}{m}\cdot \frac{\frac{C}{y}-1}{(1+\frac{y}{m})^{n+1}}
}
$$

<span style="color:blue">
    
Example: Given $D_{dollar} = 5$, what will happen to the price of the bond if the yield rises 1% (100 bps)?
$dP = -D_{dollar} \cdot dy = -5 \times 0.01 = 0.05$

The bond price will fall 5 cents if the yield rises 1%.

</span>

### Modified Duration

Modified Duration measures the percentage price change per unit (100%) yield change

$
\boxed{
\text{Modified Duration} = -\frac{1}{P} \frac{dP}{dy} = \frac{DD}{P}
}
$

<span style="color:blue">
    
Example: Given $D_{mod} = 5$, what will happen to the price of the bond if the yield rises 1% (100 bps)?
$\frac{dP}{P} = -D_{mod} \cdot dy = 5 \times 0.01 = 0.05$

The bond price will fall about 5% if the yield rises 1%.
</span>

### PV01

PV01 is defined as the change in PV of a bond due to a 1 bp change in yield

$
\boxed{
PV01 = -\frac{\partial P}{\partial y} \times 0.0001 = DD \times 0.0001
}
$

## 7. Convexity

Bond convexity is a measure of the non-linear, curved relationship between a bond's price and its yield, representing the rate of change in duration as interest rates fluctuate. We find convexity by computing the second derivative of the bond price with respect to the interest rate $y$ (or $r$).

$$
Convexity = \frac{1}{P}\frac{d^2P}{dy^2}
$$

With convexity, we can compute a second-order Taylor Expansion:

$$
\begin{aligned}
dP &\approx \frac{dP}{dy} dy \\
dP &\approx -D_{mod} dy + \frac{1}{2}\text{Convexity}\cdot dy^2
\end{aligned}
$$

For standard plain-vanilla bonds, convexity is ALWAYS positive.
- When yields fall → price rises more than duration predicts
- When yields rise → price falls less than duration predicts

A large convexity indicates the yield curve is strongly curved. The bond's price reacts more favorably to interest rate changes.
- When yields fall ($dy<0$) → the duration term make price rise while the convexity term adds extra upside.
- When yields rise ($dy>0$) → the duration term predicts a price drop while the convexity term offsets part of that loss.

Convexity increases with:
- Longer maturity
- Lower coupon
- Lower yield
